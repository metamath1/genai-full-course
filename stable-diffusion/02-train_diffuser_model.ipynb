## Step 2: Prepare a training dataset

- pytorchë¡œ ë¶€í„° MNISTë°ì´í„° ì…‹ ë‹¤ìš´ë¡œë“œ

- ì´ë¯¸ì§€ ë°°ì¹˜ë¥¼ ê·¸ë¦¬ë“œ í˜•íƒœë¡œ ë°”ê¿” í‘œì‹œí•  ë³´ì¡° í•¨ìˆ˜ ì •ì˜

```python
def show_images(x, return_type='PIL'):
    """Given a batch of images x, make a grid and convert to PIL"""

    # Map from (-1, 1) back to (0, 1)
    x = x * 0.5 + 0.5

    # (B x C x H x W) -> (C x H x W)ì¸ ì´ë¯¸ì§€ë¥¼ (B/nrow, nrow)ì¸ ê²©ìží˜•íƒœë¡œ.
    # Default nrow: 8
    # 1ì±„ë„ ì´ë¯¸ì§€ê°€ ìž…ë ¥ë˜ë„ ë°˜í™˜ì€ 3ì±„ë„ë¡œ ë¨
    grid = torchvision.utils.make_grid(x)

    #                             ch last,         (0,1) -> (0,255)
    grid_im = grid.detach().cpu().permute(1, 2, 0).clip(0, 1) * 255

    if return_type == 'PIL':
        # PIL ì´ë¯¸ì§€ë¡œ ë¦¬í„´í•˜ëŠ” ê²½ìš°
        # tensor -> ndarray -> Image
        grid_im = Image.fromarray(np.array(grid_im).astype(np.uint8))
    else:
        # numpy ì–´ë ˆì´ë¡œ ë¦¬í„´í•˜ëŠ” ê²½ìš°
        # tensor -> ndarray
        # ndarray(H,W,C)
        grid_im = np.array(grid_im).astype(np.uint8)

    return grid_im
```

```python
from torchvision import transforms

image_size = 28
batch_size = 128

# Define data augmentations
preprocess = transforms.Compose(
    [
        transforms.RandomHorizontalFlip(),  # Randomly flip (data augmentation)
        transforms.ToTensor(),  # Convert to tensor (0, 1)
        transforms.Normalize([0.5], [0.5]),  # Map to (-1, 1)
    ]
)
```

```python
dataset = torchvision.datasets.MNIST(
    root="mnist/", train=True, download=True,
    transform=preprocess)
```

```python
train_dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True, num_workers=4)
```

```python
x, y = next(iter(train_dataloader))

print('Input shape:', x.shape)
print('Labels:', y)

grid_imgs = show_images(x)
print(type(grid_imgs))

plt.imshow(grid_imgs)

plt.show()
```

## Step 3: Define the Scheduler

- ë”°ë¼ì„œ ì–´ë–¤ ìŠ¤ì¼€ì¥´ë¡œ ë…¸ì´ì¦ˆë¥¼ ì¶”ê°€í•˜ëŠ”ì§€ë„ ëª¨ë¸ì˜ í•™ìŠµì— ì˜í–¥ì„ ë¯¸ì¹¨

- DDPM([Denoising Diffusion Probabalistic Models](https://arxiv.org/abs/2006.11239))ì—ì„œ ê¸°ë³¸ì ìœ¼ë¡œ ì‚¬ìš©í•˜ëŠ” ë…¸ì´ì¦ˆ ìŠ¤ì¼€ì¥´ëŸ¬ë¥¼ ì„¸íŒ…

```python
from diffusers import DDPMScheduler

noise_scheduler = DDPMScheduler(num_train_timesteps=1000)
```

```python
xb = next(iter(train_dataloader))[0].to(device)[:8]
print("X shape:", xb.shape)
show_images(xb).resize((8 * 64, 64), resample=Image.NEAREST)
```

```python
# FutureWarning: Accessing config attribute `num_train_timesteps` directly via 'DDPMScheduler' object attribute is deprecated.
# Please access 'num_train_timesteps' over 'DDPMScheduler's config object instead,
# e.g. 'scheduler.config.num_train_timesteps'.

timesteps = torch.randint(
            0, noise_scheduler.config.num_train_timesteps, (8,)
        ).long().to(device)
print('timesteps: ', timesteps)

noise = torch.randn_like(xb)
noisy_xb = noise_scheduler.add_noise(xb, noise, timesteps)
print("Noisy X shape", noisy_xb.shape)

show_images(noisy_xb).resize((8 * 64, 64), resample=Image.NEAREST)
```

- ì œê³µëœ ìŠ¬ë¼ì´ë“œë¡œ ì‹¬ì¸µ ë¶„ì„

```python
from diffusers import UNet2DModel

model = UNet2DModel(
    sample_size=image_size,   # the target image resolution
    in_channels=1,            # the number of input channels, 3 for RGB images
    out_channels=1,           # the number of output channels
    layers_per_block=2,       # how many ResNet layers to use per UNet block
    block_out_channels=(32, 64, 64), # Roughly matching our basic unet example
    down_block_types=(
        "DownBlock2D",        # a regular ResNet downsampling block
        "AttnDownBlock2D",    # a ResNet downsampling block with spatial self-attention
        "AttnDownBlock2D",
    ),
    up_block_types=(
        "AttnUpBlock2D",
        "AttnUpBlock2D",      # a ResNet upsampling block with spatial self-attention
        "UpBlock2D",          # a regular ResNet upsampling block
      ),
)
model.to(device)
```

## Step 5: Create a Training Loop

- ëª¨ë¸ê³¼ ìŠ¤ì¼€ì¥´ëŸ¬ê°€ ìžˆì„ ë¿ í•™ìŠµì€ ì¼ë°˜ì ì¸ pytorch í•™ìŠµ ë£¨í”„ì™€ ë™ì¼

- ë‹¨ ëª¨ë¸ì— ìƒ˜í”Œ ë¯¸ë‹ˆë°°ì¹˜ë¥¼ ìž…ë ¥í•˜ê¸°ì „ì— ë¯¸ë‹ˆë°°ì¹˜ì— ë…¸ì´ì¦ˆë¥¼ ì£¼ìž…í•˜ëŠ” ê³¼ì •ì„ ì¶”ê°€

```python
# Training loop
# optimizer = torch.optim.AdamW(model.parameters(), lr=4e-4)
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-3)

epochs = 15
losses = []
print_epoch = 1

for epoch in range(epochs):
    for step, batch in enumerate(train_dataloader):
        optimizer.zero_grad()

        # sample images without noise
        clean_images = batch[0].to(device)

        # Sample noise to add to the images
        noise = torch.randn(clean_images.shape).to(clean_images.device)
        bs = clean_images.shape[0]

        # Sample a random timestep for each image
        #  0 ~ 999 íƒ€ìž… ìŠ¤íƒ­ ìˆ«ìž bsê°œ ìƒ˜í”Œë§
        # ê° ì´ë¯¸ì§€ì— ê°ê¸° ë‹¤ë¥¸ íƒ€ìž…ìŠ¤íƒ­ì„ ì§€ì •í•˜ì—¬ ê·¸ì— ë”°ë¥¸ ë…¸ì´ì¦ˆë¥¼ ì¤€ë‹¤.
        timesteps = torch.randint(
            0, noise_scheduler.config.num_train_timesteps, (bs,), device=clean_images.device
        ).long()

        # Add noise to the clean images according to the noise magnitude at each timestep
        noisy_images = noise_scheduler.add_noise(clean_images, noise, timesteps)

        # Get the model prediction
        noise_pred = model(noisy_images, timesteps).sample

        # Calculate the loss
        # ëª¨ë¸ì€ ë…¸ì´ì¦ˆ ë‚€ ì´ë¯¸ì§€ë¥¼ ìž…ë ¥ë°›ê³  ê·¸ ì´ë¯¸ì§€ì— ì¶”ê°€ëœ ë…¸ì´ì¦ˆë¥¼ ì¶œë ¥
        # ì´ ë…¸ì´ì¦ˆë¥¼ ì´ìš©í•˜ë©´ ì•½ê°„ ë…¸ì´ì¦ˆê°€ ëœ ë‚€ ì´ì „ ë‹¨ê³„ì˜ í‰ê·  ì´ë¯¸ì§€ë¥¼ ë§Œë“¤ì–´
        # ë‚¼ ìˆ˜ ìžˆë‹¤.
        # ì¦‰ VAEì²˜ëŸ¼ íŠ¹ì • të‹¨ê³„ì— ë…¸ì´ì¦ˆë‚€ ì´ë¯¸ì§€ê°€ ë“¤ì–´ì˜¤ë©´ ê·¸ ì´ë¯¸ì§€ì™€ ìƒê´€ìžˆëŠ”
        # íŠ¹ì • ë…¸ì´ì¦ˆë¥¼ ì¶œë ¥í•˜ëŠ” ê²ƒì´ê³  ì´ ë…¸ì´ì¦ˆë¥¼ ì´ìš©í•˜ë©´ t-1 ë‹¨ê³„ì˜
        # í‰ê· ì´ë¯¸ì§€ë¥¼ êµ¬í•  ìˆ˜ ìžˆê³  ì´ë¥¼ ì´ìš©í•´ t-1 ë‹¨ê³„ì˜ ë…¸ì´ì¦ˆë‚€ ì´ë¯¸ì§€ë¥¼
        # ìƒ˜í”Œë§í•  ìˆ˜ìžˆê²Œ ëœë‹¤.
        loss = F.mse_loss(noise_pred, noise)
        loss.backward(loss)
        losses.append(loss.item())

        # Update the model parameters with the optimizer
        optimizer.step()


    if (epoch + 1) % print_epoch == 0:
        loss_last_epoch = sum(losses[-len(train_dataloader) :]) / len(train_dataloader)
        print(f"Epoch:{epoch+1}, loss: {loss_last_epoch}")
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))
axs[0].plot(losses)
axs[1].plot(np.log(losses))
plt.show()
```

### Option 2: Writing a Sampling Loop

- ë‘ë²ˆì§¸ ë°©ì‹ì€ ì§ì ‘ ìƒ˜í”Œë§ ë£¨í”„ë¥¼ êµ¬í˜„


```python
# Random starting point (8 random images):
sample = torch.randn(64, 1, 28, 28).to(device)

for i, t in enumerate(noise_scheduler.timesteps):

    # Get model pred
    with torch.no_grad():
        residual = model(sample, t).sample

    # Update sample with step
    # residual: predicted epsilon
    # sample: xt
    # t: time step
    # prev_sample: x_{t-1}
    sample = noise_scheduler.step(residual, t, sample).prev_sample

show_images(sample)
```
