## URL Loader

### WebBaseLoader

- 주어진 URL 전체 내용을 다 가져옴

- https://python.langchain.com/docs/integrations/document_loaders/web_base

```python
import requests

from langchain_community.document_loaders import WebBaseLoader
```

```python
url = "https://metamath1.github.io/blog/posts/diffusion/ddpm_part1.html"
```

```python
data = requests.get(url)

# requests로는 안깨짐
data.text[-11100:]
```

```python
loader = WebBaseLoader(
    web_paths=(url,),
    encoding='utf-8' # 한글 깨지는 경우 강제 인코딩 설정
)

docs = loader.load()
```

```python
# 전체 페이지를 나누지 않고 하나로 읽어 드림
len(docs)
```

```python
# 한글 깨짐는 경우 서브클래싱으로 해결하는 방법
# https://github.com/langchain-ai/langchain/issues/9925
# 현재 버전에서는 인코딩 지정 옵션이 생김

print(docs[0].page_content[:1000])
print(docs[0].metadata)
```

```python
data = loader.load()
```

```python
print(data[0].page_content[:1000])
print(data[0].metadata)
```

# PDF Loader

- https://python.langchain.com/docs/modules/data_connection/document_loaders/pdf

```python
!pip install pypdf
```

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
# 가끔 gdown 에러 날 때
!pip install gdown==4.6.0
```

```python
!gdown 15zGjdjHS279Byb_06P1qugToEzS1ivg4
```

```python
loader = PyPDFLoader("초거대 언어모델 연구 동향_업스테이지_6000원.pdf")

# pdf 파일 페이지별로 잘라서 읽어옴
pages = loader.load_and_split()
```

```python
# 페이지 확인
len(pages)
```

```python
# 페이지 별로 잘려서 나옴
print(pages[0].page_content)
print(pages[0].metadata)
```

```python
# 한페이지가 꼭 document하나는 아님
for i in range(22):
    print(i, pages[i].metadata)
```

```python
# 11페이지는 document 두개로 나눠짐
print(pages[11].page_content)
print('-----------------------')
print(pages[12].page_content)
```

## Pandas DataFrame Loader

- https://python.langchain.com/docs/integrations/document_loaders/pandas_dataframe

```python
import pandas as pd
df = pd.read_csv("teams.csv")
```

```python
df.head()
```

```python
df['Team'] = df['Team'] + ',' + df['City']
```

```python
from langchain_community.document_loaders import DataFrameLoader
```

```python
loader = DataFrameLoader(
    df, # 데이터 프레임
    page_content_column="Team" # Document의 내용이 될 컬럼, 나머지 컬럼은 메타데이터로
)
```

```python
# 한 행이 Document 하나로 읽혀짐
data = loader.load()
data
```

## CSV Loader

- https://python.langchain.com/docs/modules/data_connection/document_loaders/csv

```python
from langchain_community.document_loaders.csv_loader import CSVLoader
```

```python
loader = CSVLoader(
    file_path='teams.csv',
    csv_args={
        'delimiter': ',',
        #'fieldnames': ['Team', 'City', 'State', 'Venue', 'League'],
        #'quotechar': '"',
    }
)
data = loader.load()
```
