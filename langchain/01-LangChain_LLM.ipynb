# OpenAI API vs LangChain

## OpenAI api

```python
import openai

client = openai.OpenAI(
    api_key=userdata.get('OPENAI_API_KEY')
)

def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0,
    )

    return response.choices[0].message.content
```

```python
get_completion("What is 1+1?")
```

# LangChain: OpenAI, ChatOpenAI

## OpenAI: Completion Model

- prompt -> completion 으로 진행되는 모델

- `text_davinch-003`, `gpt-3.5-turbo-instruct`등이 해당

```python
# from openai import OpenAI
from langchain_community.llms.openai import OpenAI

# davinci-003은 gpt3 기반 그냥 completion 모델(Legacy 모델이 되어 더 이상 사용안함)
# 대화에 특화 되어있지 않음
davinch3 = OpenAI(model_name="text-davinci-003")
davinch3.invoke('why python is the most popular language? Answer in Korean')
```

## ChatOpenAI: Chat Completion Model

- System, User->Assistant->User->Assistant->... 로 진행되는 모델


- SystemMessage: ChatGPT에게 역할을 부여하여, 대화의 맥락을 설정하는 메세지

- HumanMessage: 사용자가 ChatGPT에게 대화 또는 요청을 위해 보내는 메세지

- AIMessage: 사용자에 언어 모델이 등답한 메세지

- [참고] [Chat Completions vs. Completions](https://platform.openai.com/docs/guides/text-generation/chat-completions-vs-completions) 차이

```python
from langchain.chat_models import ChatOpenAI

chatgpt = ChatOpenAI(model_name="gpt-3.5-turbo")
answer = chatgpt.invoke("why python is the most popular language? answer in Korean")

print(answer)
```

### 매개변수

- `temperature`: 응답의 무작위성 조절 0에 가까우면 무작위성 작아지고 숫자가 커지면 무작위성 높아짐

```python
chatgpt = ChatOpenAI(model_name="gpt-3.5-turbo", temperature = 0)
answer = chatgpt.invoke("why python is the most popular language? answer in Korean")
print(answer)
```

```python
# 동일한 질문에 대해서 위와 동일한 답변
chatgpt = ChatOpenAI(model_name="gpt-3.5-turbo", temperature = 0)
answer = chatgpt.invoke("why python is the most popular language? answer in Korean")
print(answer)
```

```python
chatgpt = ChatOpenAI(model_name="gpt-3.5-turbo", temperature = 1)
answer = chatgpt.invoke("why python is the most popular language? answer in Korean")
print(answer)
```

```python
# 동일한 질문에 대해서 위와 다른 답변
chatgpt = ChatOpenAI(model_name="gpt-3.5-turbo", temperature = 1)
answer = chatgpt.invoke("why python is the most popular language? answer in Korean")
print(answer)
```

### 타이핑하듯 답변하기

```python
from langchain.callbacks.streaming_stdout import StreamingStdOutCallbackHandler

chatgpt = ChatOpenAI(
    model_name="gpt-3.5-turbo",
    streaming=True,
    callbacks=[StreamingStdOutCallbackHandler()],
    temperature = 1
)

# answer를 찍지 않아도 실시간으로 화면에 답변이 출력됨
chatgpt.invoke("why python is the most popular language? answer in Korean")
```

### 메세지 구성

```python
from langchain.schema import AIMessage, HumanMessage, SystemMessage

chatgpt = ChatOpenAI(model_name="gpt-3.5-turbo", temperature = 0)

# OpenAI API에서 하던 방식을 langchain 모델을 사용해서 활용하기
messages = [
    SystemMessage(
        content="You are a helpful assistant that translates English to Korean."
    ),
    HumanMessage(
        content="I love langchain."
    ),
]

response_langchain = chatgpt.invoke(messages)
response_langchain

# 답변은 AIMessage, OpenAI API에서는 assistant에 해당
```

```python
response_langchain.content
```

# 실습해보기

- ChatOpenAI 사용 (`model_name='gpt-3.5-turbo'`)

- messages 리스트를 다음과 같이 구성
    - SystemMessage: 당신은 공부 계획을 세워주는 스터디 플래너 머신입니다. 사용자의 공부 주제를 입력 받으면, 이를 학습하기 위한 공부 계획을 작성합니다.
    - HumanMessage: <내가 공부하고싶은 주제>

- 스트림 기능 활용

- 첫 질문에 돌아오는 `AIMessage`를 messages에 추가 후 두 번째 질문을 추가해서 답변 받기



```python
chatgpt = ChatOpenAI(temperature = 1,
                     streaming=True,
                     callbacks=[StreamingStdOutCallbackHandler()])

messages =  [
    SystemMessage(content="""
당신은 공부 계획을 세워주는 스터디 플래너 머신입니다.
사용자의 공부 주제를 입력 받으면, 이를 학습하기 위한 공부 계획을 작성합니다.
"""),
    HumanMessage(content="Large Language Model에 대해서 공부하고 싶어요.")
  ]

answer = chatgpt.invoke(messages)
```
